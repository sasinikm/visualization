```python
plt.style.use('default')
fig,(ax1, ax2) = plt.subplots(ncols=2,
nrows=1,
figsize=(7, 3)) # generate empty plot with two axes

ax1.scatter(x1,y1,
            marker = 'o',
            color = 'indigo') # modify scatterplot on axis1
ax2.bar(x2,y2,
        color = 'darkgreen') # modify bar graph on axis2
```

```python
#add 3 subplots 
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
['ax2', 'ax3']],
figsize=(7, 4))

someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
['ax2', 'ax3']],
figsize=(7, 4))

#modify subplots
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18) #modify axis1 label
someaxes["ax2"].set_xlabel('Another Label',fontsize=18) #modify axis2 label
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 ) #modify axis3 label
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
['ax2', 'ax3']],
figsize=(7, 4),
layout = "constrained") #modify subplots to fit neatly in our figure area

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label', fontsize=18)
someaxes["ax2"].set_xlabel('Another Label', fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)
```

```python
# generate sample data
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])


# define figure and axes
fig, ax = plt.subplots(figsize=(7, 3))

#multiple graphs on one axis
ax.bar(x, y1,
       color = "indigo")
ax.plot(x, y2,
        color = "red")
```

```python
y2_sd = np.std(y2) #standard deviation of our data

fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red") #plot error in a line graph

ax.errorbar(x, # x values
y2, # y values
yerr = y2_sd, # error values
fmt = "none") # actual values are not plotted
```

```python
fig, ax = plt.subplots(figsize=(7, 3))

# customize error bars
ax.plot(x, y2,
        color = "red")
ax.errorbar(x,
            y2,
            yerr = y2_sd,
            fmt = "none",
            ecolor= "indigo",
            elinewidth= 4,
            capsize = 6,
            capthick= 4 )

```

```python
fig, ax = plt.subplots(figsize=(7, 3))

# customize error bars
ax.plot(x, y2,
        color = "red")

ax.errorbar(x,
            y2,
            yerr = y2_sd,
            fmt = "none",
            ecolor= "indigo",
            elinewidth= 4,
            capsize = 6,
            capthick= 4,
            errorevery= 2) # show error bars only at specified intervals
```
